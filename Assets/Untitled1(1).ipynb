```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
data = pd.read_csv("data.csv",sep=";")
plt.plot('Time','x',data=data,label="x")
plt.plot('Time','y',data=data,label="y")
plt.plot('Time','z',data=data,label="z")
plt.xlabel("Time")
plt.ylabel("Acceleration")
plt.legend()
plt.show()
```
